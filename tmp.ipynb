```python
import cv2
from matplotlib import pyplot as plt

# 이미지 파일 불러오기
image = cv2.imread('static/images/comparsion/pancake_origin.png')
if image is None:
    print("이미지를 불러오는 데 실패했습니다. 파일 경로를 확인하세요.")
else:
    # 가우시안 블러 적용
    blurred_image = cv2.GaussianBlur(image, (9, 9), 0)

    # 블러 처리된 이미지 저장
    cv2.imwrite('static/images/comparsion/pancake_blurred.png', blurred_image)

    # OpenCV는 BGR 색상 순서를 사용하므로, 표시하기 전에 RGB로 변환
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

    # 원본 이미지와 블러 처리된 이미지를 나란히 표시
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)  # 1행 2열의 첫 번째 위치
    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)  # 1행 2열의 두 번째 위치
    plt.imshow(blurred_image_rgb)
    plt.title('Gaussian Blurred Image')
    plt.axis('off')

    plt.show()
```
